```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
df = pd.read_html('luton_town_squad_16_sept_2027.html')[0]
```

```python
goalkeepers_df = df[df['Position'] == 'GK']
```

```python
goalkeepers_df['ska_key'] = ( goalkeepers_df['Agi'] + goalkeepers_df['Ref'] )
goalkeepers_df['ska_green'] = ( goalkeepers_df['Cmd'] + goalkeepers_df['Kic'] + goalkeepers_df['1v1'] + goalkeepers_df['Ant'] + goalkeepers_df['Cnt'] + goalkeepers_df['Pos'] )
goalkeepers_df['ska_blue'] = ( goalkeepers_df['Aer'] + goalkeepers_df['Fir'] + goalkeepers_df['Han'] + goalkeepers_df['Pas'] + goalkeepers_df['TRO'] + goalkeepers_df['Dec'] + goalkeepers_df['Vis'] + goalkeepers_df['Acc'] )
goalkeepers_df['ska'] =( ( ( goalkeepers_df['ska_key'] * 5) + (goalkeepers_df['ska_green'] * 3) + (goalkeepers_df['ska_blue'] * 1) ) / 36)
goalkeepers_df.ska= goalkeepers_df.ska.round(1)
```

```python
goalkeepers_df[['Name', 'ska', 'Preferred Foot']].sort_values('ska', ascending=False)
```

```python
central_defenders_df = df[df['Position'].apply(lambda x: 'D ' in x)]
```

```python
central_defenders_df['cdd_key'] = ( central_defenders_df['Acc'] + central_defenders_df['Pac'] + central_defenders_df['Jum'] + central_defenders_df['Cmp'] )
central_defenders_df['cdd_green'] = ( central_defenders_df['Hea'] + central_defenders_df['Mar'] + central_defenders_df['Tck'] + central_defenders_df['Pos'] + central_defenders_df['Str'] )
central_defenders_df['cdd_blue'] = ( central_defenders_df['Agg'] + central_defenders_df['Ant'] + central_defenders_df['Bra'] + central_defenders_df['Cnt'] + central_defenders_df['Dec'] )
central_defenders_df['cdd'] =( ( ( central_defenders_df['cdd_key'] * 5) + (central_defenders_df['cdd_green'] * 3) + (central_defenders_df['cdd_blue'] * 1) ) / 40)
central_defenders_df.cdd= central_defenders_df.cdd.round(1)
```

```python
central_defender_filter = df['Position'].apply(lambda x: 'D ' in x)
wingback_filter = (df['Position'].apply(lambda x: 'WB' in x) |  df['Position'].apply(lambda x: '/WB' in x))
```

```python
central_defenders_df[['Name', 'Position', 'Preferred Foot','cdd',]].sort_values('cdd', ascending=False)
```

```python
central_defenders_df[['Name', 'Position', 'Preferred Foot','fbd',]].sort_values('fbd', ascending=False)
```

```python
full_backs_df = df[wingback_filter | central_defender_filter]
```

```python
full_backs_df['fbd_key'] = ( full_backs_df['Acc'] + full_backs_df['Pac'] + full_backs_df['Sta'] + full_backs_df['Wor'] )
full_backs_df['fbd_green'] = ( full_backs_df['Mar'] + full_backs_df['Tck'] + full_backs_df['Ant'] + full_backs_df['Cnt'] + full_backs_df['Pos'] + full_backs_df['Pos'] )
full_backs_df['fbd_blue'] = ( full_backs_df['Cro'] + full_backs_df['Pas'] + full_backs_df['Dec'] + full_backs_df['Tea'] )
full_backs_df['fbd'] =( ( ( full_backs_df['fbd_key'] * 5) + (full_backs_df['fbd_green'] * 3) + (full_backs_df['fbd_blue'] * 1) ) / 42)
full_backs_df.fbd= full_backs_df.fbd.round(1)
```

```python
full_backs_df[['Name', 'Position', 'Preferred Foot','fbd',]].sort_values('fbd', ascending=False)
```

```python
full_backs_df['iwbs_key'] = ( full_backs_df['Acc'] + full_backs_df['Pac'] + full_backs_df['Sta'] + full_backs_df['Wor'] )
full_backs_df['iwbs_green'] = ( full_backs_df['Fir'] + full_backs_df['Pas'] + full_backs_df['Tck'] + full_backs_df['Cmp'] + full_backs_df['Dec'] + full_backs_df['Tea'] )
full_backs_df['iwbs_blue'] = ( full_backs_df['Mar'] + full_backs_df['Tec'] + full_backs_df['Ant'] + full_backs_df['Cnt'] + full_backs_df['OtB'] + full_backs_df['Pos'] + full_backs_df['Vis'] + full_backs_df['Agi'] )
full_backs_df['iwbs'] =( ( ( full_backs_df['iwbs_key'] * 5) + (full_backs_df['iwbs_green'] * 3) + (full_backs_df['iwbs_blue'] * 1) ) / 46)
full_backs_df.iwbs= full_backs_df.iwbs.round(1)
```

```python
full_backs_df[['Name', 'Position', 'Preferred Foot','iwbs',]].sort_values('iwbs', ascending=False)
```

```python
midfield_filter = (df['Position'].apply(lambda x: 'M ' in x) | df['Position'].apply(lambda x: 'M/' in x) | df['Position'].apply(lambda x: '/M' in x))
defensive_mid_filter = (df['Position'].apply(lambda x: 'DM' in x) | df['Position'].apply(lambda x: 'DM/' in x) | df['Position'].apply(lambda x: '/DM' in x))
attacking_mid_filter = (df['Position'].apply(lambda x: 'AM ' in x) | df['Position'].apply(lambda x: 'AM/' in x) | df['Position'].apply(lambda x: '/AM' in x))
```

```python
defensive_mid_df = df[defensive_mid_filter]
```
